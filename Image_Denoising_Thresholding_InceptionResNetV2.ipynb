**Coding/pipeline to convert the original image into thresholded image to get only important features**

```python
#for training dataset

class segmentation:
  def __init__(self,directory_path):
    self.directory_path=directory_path

  def image_paths(self):
    l=[]
    for i in (["{:02}".format(i) for i in range(1,22)]):
      for j in (["{:02}".format(i) for i in range(1,33)]):
        l.append(os.path.join(self.directory_path,i,'EMDS5-g{}-{}.png'.format(i,j)))
    return l

  def image_read(self):
    l2=[]
    for i in  self.image_paths():
      img=cv2.imread(i,0)
      l2.append(img)
    return l2 

  def blurring_image(self):
     l3=[]
     for i in self.image_read():
       blur_image=cv2.medianBlur(i,5)
       l3.append(blur_image)
     return l3 
  def thresholding(self):
    l4=[]
    for i in self.blurring_image():
      thresholded_image=cv2.adaptiveThreshold(i,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,11,2)  
      l4.append(thresholded_image)
    return (l4)
    

  def making_dir(self):
    if not os.path.exists('/content/drive/MyDrive/Colab Notebooks/EMDS5-SEGMENTED-IMAGE'):
      os.mkdir('/content/drive/MyDrive/Colab Notebooks/EMDS5-SEGMENTED-IMAGE')
    for m in  ["{:02}".format(i) for i in range(1,3)]:
      l5={"01":'Train',"02":'Valid'}
      if not os.path.exists(os.path.join('/content/drive/MyDrive/Colab Notebooks/EMDS5-SEGMENTED-IMAGE',l5[m])):
        os.mkdir(os.path.join('/content/drive/MyDrive/Colab Notebooks/EMDS5-SEGMENTED-IMAGE',l5[m])) 
      for n in (["{:02}".format(i) for i in range(1,22)]):
        if not os.path.exists(os.path.join('/content/drive/MyDrive/Colab Notebooks/EMDS5-SEGMENTED-IMAGE/',l5[m],n)) :
          os.mkdir(os.path.join('/content/drive/MyDrive/Colab Notebooks/EMDS5-SEGMENTED-IMAGE/',l5[m],n)) 

  def saving_segmented_image(self):
    l6=[]
    for i in self.thresholding():
      l6.append(i)
    for j in range(len(l6)):  
      if j<32:
        cv2.imwrite('/content/drive/MyDrive/Colab Notebooks/EMDS5-SEGMENTED-IMAGE/Train/01'+'/'+str(j)+'.png',l6[j])
      elif 32<=j<64:
        cv2.imwrite('/content/drive/MyDrive/Colab Notebooks/EMDS5-SEGMENTED-IMAGE/Train/02'+'/'+str(j)+'.png',l6[j])
      elif 64<=j<96:
        cv2.imwrite('/content/drive/MyDrive/Colab Notebooks/EMDS5-SEGMENTED-IMAGE/Train/03'+'/'+str(j)+'.png',l6[j])
      elif 96<=j<128:
        cv2.imwrite('/content/drive/MyDrive/Colab Notebooks/EMDS5-SEGMENTED-IMAGE/Train/04'+'/'+str(j)+'.png',l6[j])
      elif 128<=j<160:
        cv2.imwrite('/content/drive/MyDrive/Colab Notebooks/EMDS5-SEGMENTED-IMAGE/Train/05'+'/'+str(j)+'.png',l6[j])
      elif 160<=j<192:
        cv2.imwrite('/content/drive/MyDrive/Colab Notebooks/EMDS5-SEGMENTED-IMAGE/Train/06'+'/'+str(j)+'.png',l6[j])
      elif 192<=j<224:
        cv2.imwrite('/content/drive/MyDrive/Colab Notebooks/EMDS5-SEGMENTED-IMAGE/Train/07'+'/'+str(j)+'.png',l6[j]) 
      elif 224<=j<256:
        cv2.imwrite('/content/drive/MyDrive/Colab Notebooks/EMDS5-SEGMENTED-IMAGE/Train/08'+'/'+str(j)+'.png',l6[j])
      elif 256<=j<288:
        cv2.imwrite('/content/drive/MyDrive/Colab Notebooks/EMDS5-SEGMENTED-IMAGE/Train/09'+'/'+str(j)+'.png',l6[j])          
      elif 288<=j<320:
        cv2.imwrite('/content/drive/MyDrive/Colab Notebooks/EMDS5-SEGMENTED-IMAGE/Train/10'+'/'+str(j)+'.png',l6[j])
      elif 320<=j<352:
        cv2.imwrite('/content/drive/MyDrive/Colab Notebooks/EMDS5-SEGMENTED-IMAGE/Train/11'+'/'+str(j)+'.png',l6[j])
      elif 352<=j<384:
        cv2.imwrite('/content/drive/MyDrive/Colab Notebooks/EMDS5-SEGMENTED-IMAGE/Train/12'+'/'+str(j)+'.png',l6[j])
      elif 384<=j<416:
        cv2.imwrite('/content/drive/MyDrive/Colab Notebooks/EMDS5-SEGMENTED-IMAGE/Train/13'+'/'+str(j)+'.png',l6[j])                    
      elif 416<=j<448:
        cv2.imwrite('/content/drive/MyDrive/Colab Notebooks/EMDS5-SEGMENTED-IMAGE/Train/14'+'/'+str(j)+'.png',l6[j])
      elif 448<=j<480:
        cv2.imwrite('/content/drive/MyDrive/Colab Notebooks/EMDS5-SEGMENTED-IMAGE/Train/15'+'/'+str(j)+'.png',l6[j])
      elif 480<=j<512:
        cv2.imwrite('/content/drive/MyDrive/Colab Notebooks/EMDS5-SEGMENTED-IMAGE/Train/16'+'/'+str(j)+'.png',l6[j])
      elif 512<=j<544:
        cv2.imwrite('/content/drive/MyDrive/Colab Notebooks/EMDS5-SEGMENTED-IMAGE/Train/17'+'/'+str(j)+'.png',l6[j])
      elif 544<=j<576:
        cv2.imwrite('/content/drive/MyDrive/Colab Notebooks/EMDS5-SEGMENTED-IMAGE/Train/18'+'/'+str(j)+'.png',l6[j])        
      elif 576<=j<608:
        cv2.imwrite('/content/drive/MyDrive/Colab Notebooks/EMDS5-SEGMENTED-IMAGE/Train/19'+'/'+str(j)+'.png',l6[j])
      elif 608<=j<640:
        cv2.imwrite('/content/drive/MyDrive/Colab Notebooks/EMDS5-SEGMENTED-IMAGE/Train/20'+'/'+str(j)+'.png',l6[j])
      elif 640<=j<672:
        cv2.imwrite('/content/drive/MyDrive/Colab Notebooks/EMDS5-SEGMENTED-IMAGE/Train/21'+'/'+str(j)+'.png',l6[j])


```

```python
import numpy as np 
import matplotlib.pyplot as plt
import glob
import cv2

from keras.models import Model, Sequential
from keras.layers import Dense, Flatten, Conv2D, MaxPooling2D
#from keras.layers.normalization import BatchNormalization
import os
import seaborn as sns
from keras.applications.inception_resnet_v2 import InceptionResNetV2
from sklearn.model_selection import GridSearchCV
```

**Data Loading (Denoised and thresholded image)**

```python
PATH='/content/drive/MyDrive/Colab Notebooks/EMDS5-SEGMENTED-IMAGE'
#PATH_Test='/content/drive/MyDrive/Colab Notebooks/EMDS5-Original'


train_dir = os.path.join(PATH, 'Train')
test_dir= os.path.join(PATH, 'Valid')



BATCH_SIZE = 1
SIZE=299
IMG_SIZE = (SIZE,SIZE)

train_dataset = tf.keras.utils.image_dataset_from_directory(train_dir,
                                                            shuffle=True,
                                                            batch_size=BATCH_SIZE,
                                                            image_size=IMG_SIZE)
```

```python
validation_dataset = tf.keras.utils.image_dataset_from_directory(test_dir,
                                                                 shuffle=True,
                                                                 batch_size=BATCH_SIZE,
                                                                 image_size=IMG_SIZE)
```

**Data prepration for test and train**

```python
train_batch_img=[]
train_batch_labels=[]
for img,labels in train_dataset:
  train_batch_img.append(img)
  train_batch_labels.append(labels)
```

```python
train_images=[]
train_labels=[]
for i in range(672):
  train_images.append(train_batch_img[i][0])
for j in range(672):
  train_labels.append(train_batch_labels[j][0])  
```

```python
#Convert lists to arrays        
train_images = np.array(train_images)
train_labels = np.array(train_labels)
```

```python
test_batch_img=[]
test_batch_labels=[]
for img,labels in validation_dataset:
  test_batch_img.append(img)
  test_batch_labels.append(labels)
```

```python
test_images=[]
test_labels=[]
for i in range(168):
  test_images.append(test_batch_img[i][0])
for j in range(168):
  test_labels.append(test_batch_labels[j][0]) 
```

```python
#Convert lists to arrays                
test_images = np.array(test_images)
test_labels = np.array(test_labels)
```

```python
#Encode labels from text to integers.
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
le.fit(test_labels)
test_labels_encoded = le.transform(test_labels)
le.fit(train_labels)
train_labels_encoded = le.transform(train_labels)
```

```python
#Split data into test and train datasets (already split but assigning to meaningful convention)
x_train, y_train, x_test, y_test = train_images, train_labels_encoded, test_images, test_labels_encoded
```

```python
# Normalize pixel values to between 0 and 1
x_train, x_test = x_train / 255.0, x_test / 255.0
```

```python
from tensorflow.keras.utils import to_categorical
```

```python
#One hot encode y values for neural network. 
#from keras.utils import to_categorical
y_train_one_hot = to_categorical(y_train)
y_test_one_hot = to_categorical(y_test)
```

**Loading the Pretrained model and extracting the features of Thresholded image **

```python
from keras.applications import inception_resnet_v2
#Load model wothout classifier/fully connected layers
Resnet_model = InceptionResNetV2(weights='imagenet', include_top=True, input_shape=(299, 299, 3))
```

```python
#Make loaded layers as non-trainable. This is important as we want to work with pre-trained weights
for layer in Resnet_model.layers:
	layer.trainable = False
```

```python
Resnet_model.summary()  #Trainable parameters will be 0
```

**Applying RF on the feature extracted by the Transfer Learning **

**Random Forest**

```python
#Now, let us use features from convolutional network for RF
feature_extractor=Resnet_model.predict(x_train)
```

```python
features = feature_extractor.reshape(feature_extractor.shape[0], -1)
```

```python
X_for_RF = features #This is our X input to RF
```

```python
#RANDOM FOREST
from sklearn.ensemble import RandomForestClassifier
RF_model = RandomForestClassifier(n_estimators = 30, random_state = 42)
```

```python
# Train the model on training data
RF_model.fit(X_for_RF, y_train) #For sklearn no one hot encoding
```

```python
#Send test data through same feature extractor process
X_test_feature = Resnet_model.predict(x_test)
X_test_features = X_test_feature.reshape(X_test_feature.shape[0], -1)
```

```python
#Now predict using the trained RF model. 
prediction_RF = RF_model.predict(X_test_features)
#Inverse le transform to get original label back. 
prediction_RF = le.inverse_transform(prediction_RF)
```

```python
#Print overall accuracy
from sklearn import metrics
print ("Accuracy = ", metrics.accuracy_score(test_labels, prediction_RF))
```
